### Análisis y transformación de las variables Provincia, Partido y Población:

```python
# Cantidad de provincias
len(data.Provincia.value_counts())
```

```python
# Cantidad de registros por provincia
data.Provincia.value_counts()
```

```python
# Se calcula la cantidad de población por provincia 

data_poblacion_sum = data.groupby('Provincia').sum('Poblacion')
data_poblacion_sum.reset_index(inplace=True)
poblacion_provincia = data_poblacion_sum[['Provincia', 'Poblacion']]
poblacion_provincia
```

```python
# Se genera un diccionario con las provincias como claves y la cantidad de poblacion como valores. Esto se realiza para poder incorporar al DF base la cantidad de poblacion de cada provincia.  
diccionario_poblacion = dict(zip(poblacion_provincia['Provincia'], poblacion_provincia['Poblacion']))
diccionario_poblacion
```

```python
# Visualizamos las 10 provincias 
data_poblacion_top10 = poblacion_provincia.sort_values(by = 'Poblacion', ascending=False).head(10)
data_poblacion_top10.reset_index(inplace=True)

data_poblacion_top10 = data_poblacion_top10.rename(columns={'Poblacion': 'Cantidad de Poblacion'})

data_poblacion_top10
```

```python
# Gráfico de las 10 provincias con mayor población.

provincias_sup = data_poblacion_top10['Provincia']
poblacion_sup = data_poblacion_top10['Cantidad de Poblacion']

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  
plt.bar(provincias_sup, poblacion_sup, color='skyblue') 
plt.title('Población por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Población')

plt.yscale('linear')
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.0f}M')
plt.gca().yaxis.set_major_formatter(formatter)


plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
# Visualizamos las 10 provincias 
data_poblacion_inf10 = poblacion_provincia.sort_values(by = 'Poblacion').head(10)
data_poblacion_inf10.reset_index(inplace=True)

data_poblacion_inf10 = data_poblacion_inf10.rename(columns={'Poblacion': 'Cantidad de Poblacion'})

data_poblacion_inf10
```

```python
# Gráfico de las 10 provincias con mayor población.

provincias_inf = data_poblacion_inf10['Provincia']
poblacion_inf = data_poblacion_inf10['Cantidad de Poblacion']

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))  
plt.bar(provincias_inf, poblacion_inf, color='skyblue') 
plt.title('Población por Provincia')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de Poblacion')

plt.yscale('linear')
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.0f}M')
plt.gca().yaxis.set_major_formatter(formatter)


plt.xticks(rotation=45)
plt.tight_layout()  
plt.show()
```

```python
# Importar bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Cargar archivo CSV
data = pd.read_csv('ConectividadalserviciodeInternet_edit.csv', encoding='latin-1')

# Reemplazar valores de 'SI' y '--' por 1 y 0 respectivamente
data = data.replace({'SI': 1, '--': 0})

# Visualizar la distribución de la población
plt.figure(figsize=(10,6))
sns.histplot(data['Poblacion'], bins=30)
plt.title('Distribución de la Población')
plt.show()

# Visualizar la correlación entre las variables
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlación entre las Variables')
plt.show()
```

** Tratamiento los datos anómalos**

```python



# Ver las primeras filas del conjunto de datos
print(data.head())

# Ver la información general del conjunto de datos
print(data.info())

# Ver estadísticas descriptivas del conjunto de datos
print(data.describe())

# Visualizar la distribución de las variables numéricas
for column in data.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(5, 5))
    sns.histplot(data[column])
    plt.title(f'Distribución de {column}')
    plt.show()

# Tratar los datos anómalos
# Supongamos que 'columna' es una columna numérica del conjunto de datos
columna = 'Poblacion'
Q1 = data[columna].quantile(0.25)
Q3 = data[columna].quantile(0.75)
IQR = Q3 - Q1

# Definir los límites para los valores atípicos
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

# Reemplazar los valores atípicos por los límites
data[columna] = np.where(data[columna] < lower_limit, lower_limit, data[columna])
data[columna] = np.where(data[columna] > upper_limit, upper_limit, data[columna])

# Verificar el tratamiento de los datos anómalos
plt.figure(figsize=(5, 5))
sns.histplot(data[columna])
plt.title(f'Distribución de {columna} después del tratamiento de los datos anómalos')
plt.show()
```

**Inspección gráfica de los estadísticos**

Observar de manera gráfica si se aprecia relación entre "y" y las "X" (ekis).

```python
# Supongamos que 'y' es tu variable dependiente y 'X1', 'X2', 'X3' son tus variables independientes
y = 'Poblacion'
X = ['ADSL', 'Cablemodem', 'Fibra optica']

for x in X:
    plt.figure(figsize=(5, 5))
    plt.scatter(data[x], data[y])
    plt.xlabel(x)
    plt.ylabel(y)
    plt.title(f'Relación entre {y} y {x}')
    plt.show()
```
